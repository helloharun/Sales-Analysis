## Exploratory Data Analysis

#### Gender

```python
plt.figure(figsize=(10,7))

# defining Seaborn color palette to use 
palette_color = sns.color_palette('tab10') 

#Counting the occurrences of each gender
gender_counts = df['Gender'].value_counts()

labels = ['Female','Male']


#Defining exploding pie
explode = [0.1, 0]
    
# plotting data on chart 
plt.pie(x=gender_counts, labels=labels, colors=palette_color, autopct='%.0f%%', explode=explode) 
plt.title('Gender Distribution')
plt.legend()  

# displaying chart 
plt.show() 
```

> From above pie chart, we can observe that the female customers are over twice than the male customers. In other words, the company's products are mostly preferred by the female customers. 

```python
gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
gen
```

```python
user_count = df.groupby('Gender')['Gender'].count().reset_index(name='User count')
user_count
```

```python
merged_gen = gen.merge(user_count, on='Gender')
merged_gen
```

```python
merged_gen['Avg txn per user'] = merged_gen['Amount']/merged_gen['User count']
merged_gen
```

> Looking at the "merged_gen" table, we can see that there are more female customers (7,832) than male customers (3,407). This means our products are more popular among females and we're doing a good job of keeping them (retaining) as customers. However, it also shows that we have a chance to grow our business by attracting more male customers. Additionally, the table shows that female customers tend to spend more money on average (9,491) than male customers (9,367). To meet our goals, we should focus on getting more male customers and finding ways to encourage female customers to buy from us more often. We can do this by understanding what products our customers like and looking for ways to improve and grow our business.

#### Age

```python
df.columns
```

```python
plt.figure(figsize=(16,10))
ax = sns.countplot(data=df, x='Age_Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_xlabel('Age Group')
plt.show()
```

```python
#Total spending by age group
spending_by_age_group = df.groupby(['Age_Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

spending_by_age_group
```

```python
plt.figure(figsize=(20,10))
ax = sns.barplot(x='Age_Group', y='Amount', data=spending_by_age_group)
for bars in ax.containers:
    ax.bar_label(bars, [f"{bar.get_height():.2f}" for bar in bars])
ax.set_title('Spending by Age Group')
ax.set_xlabel('Age Group')

plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: '{:,.2f}'.format(x)))

plt.show()
```

> The first graph shows that people between 26 and 35 years old spend the most money. People between 0 and 17 years old spend the least. It also shows that most customers are women.

> The latter graph shows that people between 26 and 35 years old spend the most money, with a total of around 42.61 million. The youngest group, those aged 0-17, spends the least, at only 2.70 million approximately. The graph shows spending decreases as age increases. To understand spending better, we should consider things like income, job, and credit score, along with age.

## State

```python
df.columns
```

```python
#Top 10 states with their respective total orders

order_by_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
order_by_state
```

```python
plt.figure(figsize=(18,8))
ax = sns.barplot(x='State',y='Orders',data=order_by_state)

for bars in ax.containers:
    ax.bar_label(bars)

    ax.set_title('Total order by state')
plt.show()
```

> The chart shows the total number of orders by state. Uttar Pradesh has the highest number of orders which is 4813, followed by Maharashtra and Karnataka, 3811 and 3241 respectively. Gujarat has the lowest number of orders i. e. 1070 only. The chart is a bar chart, with the height of each bar representing the number of orders in each state.

```python
# Top 10 states by total revenue 
revenue_by_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
revenue_by_state
```

```python
plt.figure(figsize=(18,8))
ax = sns.barplot(x='State',y='Amount',data=revenue_by_state)

for bars in ax.containers:
    ax.bar_label(bars, [f"{bar.get_height():.2f}" for bar in bars])

    ax.set_title('Revenue by state')

    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x)))
plt.show()
```

> The bar chart shows that the Uttar Pradesh has the highest revenue, followed by Maharashtra and Karnataka; 19.37 million, 14.43 million, and 13.52 million respectively. And the Gujarat is having the lowest revenue which is about 3.9 million only.

## Product Category

```python
plt.figure(figsize=(25,8))
ax = sns.countplot(df, x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
ax.set_title('Product categories count')
plt.show()
```

> The graph shows the amount of revenue by product category. The category with the highest sales is "Auto" with over 20,000 units sold. The second highest is "Tupperware", followed by "Hand & Power Tools". Sales decline steadily from there, with "Office" having the lowest sales at just over 700 units. It appears that there are much higher sales for product categories that would be considered "home" or "household" related.
